```python
#Creamos un dataframe de una sola columna
df = pd.DataFrame(index=[9, 8, 5, 4, 3, 6],
                  columns=['Frecuencia_Abs'],data=[15, 12, 27, 24, 14, 49])

print(df)
df=df.sort_index()
df
```

```python
# @title Frecuencia_Abs

from matplotlib import pyplot as plt
df['Frecuencia_Abs'].plot(kind='line', figsize=(8, 4), title='Frecuencia_Abs')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
print(df.dtypes)
```

```python
#Frecuencia Absoluta Acumulada
Frec_Abs_val=df['Frecuencia_Abs'].values
Frec_Abs_val
acum = []
valor_acum = 0
for i in Frec_Abs_val:
  valor_acum = valor_acum + i
  acum.append(valor_acum)

df['Frec_Abs_Acum']=acum
df
```

```python
#Creamos la Frecuencia Relativa#

Total = df['Frecuencia_Abs'].sum()
df["Frec_Rel"]=100*(df["Frecuencia_Abs"]/Total)
df
```

```python
#Frecuencia Relativa Acumulada
Frec_rel_val=df['Frec_Rel'].values
Frec_rel_val
acum = []
valor_acum = 0
for i in Frec_rel_val:
  valor_acum = valor_acum + i
  acum.append(valor_acum)

df['Frec_Rel_Acum']=acum
df
```

```python
# Datos de ejemplo
data = {
    'Categoría': ['A', 'B', 'C', 'D', 'E'],
    'Valor': [10, 25, 15, 30, 20]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Gráfico de barras
plt.figure(figsize=(8, 4))
plt.bar(df['Categoría'], df['Valor'])
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.title('Gráfico de Barras')
plt.show()

# Gráfico de sectores
plt.figure(figsize=(6, 6))
plt.pie(df['Valor'], labels=df['Categoría'], autopct='%1.1f%%')
plt.title('Gráfico de Sectores')
plt.show()

# Gráfico de líneas
data = {
    'Año': [2010, 2011, 2012, 2013, 2014],
    'Ventas': [100, 120, 90, 150, 130]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
plt.plot(df['Año'], df['Ventas'], marker='o', linestyle='-', color='b')
plt.xlabel('Año')
plt.ylabel('Ventas')
plt.title('Gráfico de Líneas')
plt.grid(True)
plt.show()

# Gráfico de dispersión
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 4, 3, 5, 6]
}
df = pd.DataFrame(data)
plt.figure(figsize=(6, 4))
plt.scatter(df['X'], df['Y'], color='r', marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersión')
plt.grid(True)
plt.show()

# Histograma
data = {
    'Datos': [3, 2, 4, 3, 5, 2, 1, 7, 6, 4, 3, 2]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
plt.hist(df['Datos'], bins=5, edgecolor='k')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.grid(True)
plt.show()

# Diagrama de caja
data = {
    'Grupo': ['A', 'A', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D', 'D'],
    'Valor': [10, 15, 20, 22, 30, 32, 33, 40, 45, 50, 60]
}
df = pd.DataFrame(data)
plt.figure(figsize=(6, 4))
plt.boxplot(df['Valor'], vert=False)
plt.xlabel('Valor')
plt.title('Diagrama de Caja')
plt.grid(True)
plt.show()
```
