# **Medidas de tendencia central**

Son medidas que nos ayudan a resumir una gran cantidad de información en un solo numero

## **Media:**
La media, también conocida como promedio aritmético, es una medida estadística que se utiliza para representar el valor central de un conjunto de datos. Se calcula sumando todos los valores de los datos y luego dividiendo esa suma por la cantidad total de valores. Es una de las medidas de tendencia central más comunes. Además puede ser susceptible a valores atípicos

**Media** = $$ \frac{1}{N} \sum_{i=1}^N x_i  $$




```python
datos = [ 2, 4, 4, 4, 5, 5, 7, 9 ,80]
media = sum(datos)/len(datos)
print(media)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #Graficar
%matplotlib inline
```

```python
# Ejemplo 1: Consumo Promedio de Energía Residencial
consumo_hogares = pd.DataFrame(columns=['Consumo (KW)'],data= [ 2, 4, 4, 4, 5, 5, 7, 9 ,80])
print(consumo_hogares)
```

```python
#Media
media_consumo =consumo_hogares['Consumo (KW)'].mean()
print("Media de consumo de energía:", media_consumo)
```

## **Mediana:**

La mediana es otra medida de tendencia central que representa el valor medio en un conjunto de datos ordenado. A diferencia de la media, la mediana no se ve afectada por valores extremos o atípicos.


###Para calcular la mediana, se siguen los siguientes pasos:

1. Ordene los datos de menor a mayor.

2. Si la cantidad de datos es impar, la mediana es el valor que está en el medio del conjunto.

3. Si la cantidad de datos es par, la mediana es el promedio de los dos valores centrales.

### **Mediana(n=impar)** = $$ x_{(n+1)/2}^{\text{ordered}} $$

### **Mediana(n=par)** = $$ \frac{x_{(n/2)}^{\text{ordered}} + x_{(n/2)+1}^{\text{ordered}}}{2} $$

```python
datos = [ 4, 4, 4, 5, 5, 7, 9, 2 ,80]

#step 1:Sort the data
datos.sort()
print(datos)
```

```python
#step 2: Find the median

n = len(datos)

if n % 2 == 1:
  # If odd, take the middle element
  mediana = datos[n//2]
else:
  # If even, take the average of the two middle elements
  mediana = (datos[n//2 - 1] + datos[n//2]) / 2

print("Mediana:", mediana)
```

```python
mediana_consumo =consumo_hogares['Consumo (KW)'].median()
print(f"Mediana de consumo de energía: {mediana_consumo}")
```

```python
# Ejemplo 2: Mediana de Cargas de Potencia
cargas_potencia = pd.DataFrame(columns=['Carga'],data=[50, 60, 70, 55, 80, 65, 75, 70, 90, 110])
print(cargas_potencia)
#mediana
mediana_cargas = cargas_potencia['Carga'].median()
print("Mediana de cargas de potencia:", mediana_cargas)
```

## **Moda:**
Es el dato que mas se repite, la moda no aplica para datos numéricos continuos

### **Moda** = $$ x_k $$ donde $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$

```python
from collections import Counter

data = [1, 2, 2, 3, 3, 3, 4, 3, 4, 4]

#Count the frequency of each element
counter = Counter(data)
print(counter)
```

```python
#Find the maximum frequency
max_count = max(counter.values())
print(max_count)
```

```python
print(counter.items())
```

```python
#Extraxt all elements with the maximum frequency (modes)

modes = [key for key, count in counter.items() if count == max_count]

print(f'the mode(s) are {modes}')
```

```python
# Ejemplo 3: Moda en Patrones de Uso Horario
uso_horario = pd.DataFrame(columns=['Uso'],data=[8, 7, 9, 8, 9, 8, 7, 10, 9, 8, 7, 8])
print(uso_horario)
moda_uso_horario = uso_horario['Uso'].mode()
print("Moda en patrones de uso horario:", moda_uso_horario)
```

##**Ejercicio 4: Variación de Tensión**

Se tienen una serie de datos que representan la variación de tensión en un sistema eléctrico durante un día. Calcule la media, mediana y moda de estos datos para entender la tendencia central de la variación de tensión en ese período.

```python
variacion_tension = pd.DataFrame(columns=['tension'],data=[120, 125, 123, 118, 122, 121, 120, 124, 123])
print(variacion_tension)
media_tension = variacion_tension['tension'].mean()
mediana_tension = variacion_tension['tension'].median()
moda_tension = variacion_tension['tension'].mode() [0]
print('\n')
print(variacion_tension.tension.sort_values())
print("Media de variación de tensión:", media_tension)
print("Mediana de variación de tensión:", mediana_tension)
print("Moda de variación de tensión:", moda_tension)
```

# **Tablas de Frecuencias**

Son herramientas utilizadas en estadística para organizar y resumir datos categóricos o cualitativos. Estas tablas permiten contar y presentar de manera ordenada la frecuencia con la que aparece cada categoría en un conjunto de datos. En otras palabras, muestran cuántas veces ocurre cada valor en una variable categórica.

```python
df = pd.DataFrame({'Persona':['Juan', 'María', 'Pedro', 'Laura','Carlos','Ana'],
                   'Tipo_Mascota':['Perro','Gato','Perro','Pez','Perro','Gato']})
df
```

##**Tabla de Frecuencia Absoluta**

```python
Agrupado=(df.groupby("Tipo_Mascota")
          .agg(Frecuencia_Abs=("Tipo_Mascota", "count")))
#Se ordenan los datos
Agrupado = Agrupado.sort_values('Frecuencia_Abs')
Agrupado.reset_index()
```

##**Tabla de Frecuencia Absoluta + Acumulada**

```python
#Creamos la Frecuencia Absoluta Acumulada
Frec_Abs_val=Agrupado['Frecuencia_Abs'].values
print(Frec_Abs_val)
acum = []
valor_acum = 0
for i in Frec_Abs_val:
  valor_acum = valor_acum + i
  acum.append(valor_acum)

# Creamos la columna Frec_Abs_Acum y se le asigna los valores acumulados
Agrupado['Frec_Abs_Acum']=acum
Agrupado
```

##**Tabla de Frecuencia Absoluta + Acumulada + Relativa**

```python
#Creamos la Frecuencia Relativa#

Total = Agrupado['Frecuencia_Abs'].sum()
print(Total)

# Creamos la columna Frec_Rel y se le asigna los valores Frecuencia_Abs dividido el total
Agrupado["Frec_Rel"]=100*(Agrupado["Frecuencia_Abs"]/Total)
Agrupado
```

```python
#Frecuencia Relativa Acumulada

Frec_rel_val=Agrupado['Frec_Rel'].values
acum = []
valor_acum = 0
for i in Frec_rel_val:
  valor_acum = valor_acum + i
  acum.append(valor_acum)

# Creamos la columna Frec_Rel_Acum y se le asigna los valores guardados en acum
Agrupado['Frec_Rel_Acum']=acum
Agrupado
```

## **Tabla de Frecuencias para el Ejemplo**

```python
#Creamos un dataframe de una sola columna
simple_df = pd.DataFrame(columns=['Dato'],data=[1,5,3,5,4,4,5,3,1,5])
simple_df
```

```python
#Crear tabla de frecuencias
Agrupado=(simple_df.groupby("Dato")
          .agg(frequencia=("Dato", "count"))
          .reset_index())
Agrupado
```

```python
#Graficar tabla de frecuencias (histograma)
simple_df.Dato.plot.hist()
```

```python
#Graficar tabla de frecuencias (Diagrama de Barras)
plt.bar(Agrupado['Dato'],Agrupado['frequencia'])
```
