## **Varianza y Desviación Estandar**

## **Varianza:**

La varianza de una población de tamaño N es el promedio de los cuadrados de las
desviaciones con respecto a la media. Al ser un parámetro de la población se usa una letra griega para designarla

$$Varianza = \frac{\sum(X_i-\bar{X})^2}{n-1} $$

```python
datas= [9,5,7,4,3,4,6,10,4,6,2,8,20]

media = sum(datas) / len(datas)
print(media)

varianza = sum((x - media) ** 2 for x in datas) / (len(datas) - 1)

print(f"La varianza es: {varianza}")
```

## **Desviación:**

También conocida como desvío típico y representada de manera
abreviada por la letra griega minúscula sigma σ o la letra latina s, así
como por las siglas SD (standard deviation)) es una medida que se
utiliza para cuantificar la variación o la dispersión de un conjunto de
datos numéricos.

$$Varianza = \sqrt{ \frac{\sum(X_i-\bar{X})^2}{n-1}} $$

```python
# Lista de datos
datos = [9,5,7,4,3,4,6,10,4,6,2,8,20]

# Calcular la media
media = sum(datos) / len(datos)

# Calcular la varianza manualmente
varianza_manual = sum((x - media) ** 2 for x in datos) / (len(datos) - 1)

# Calcular la desviación estándar manualmente
desviacion_estandar_manual = varianza_manual ** 0.5

print(f"La varianza calculada manualmente es: {varianza_manual}")
print(f"La desviación estándar calculada manualmente es: {desviacion_estandar_manual}")
```

```python
# Varianza
var = df["Nota"].var()
std_1=np.sqrt(var)
print("""
      Varianza: %f
      Desviación: %f
""" % (var,std_1))

print('Desviación Estandar')
std = df["Nota"].std()
print(std)
assert(std_1 == std)
```

```python
# El coeficiente de variación (CV) es una medida estadística que se utiliza para
#evaluar la variabilidad relativa de un conjunto de datos en relación con su media (promedio).
import scipy.stats as ss
cv = df["Nota"].std() / df["Nota"].mean()
cv2 = ss.variation(df["Nota"])
#assert(cv == cv2)
print(cv)
print(cv2)
```

## **Cuartiles**

```python
# Mostrar Cuartiles, Q2 es la mediana
median = df['Nota'].median()
Q1 = df['Nota'].quantile(q=0.25)
Q3 = df['Nota'].quantile(q=0.75)
min_val = df['Nota'].quantile(q=0)
max_val = df['Nota'].quantile(q=1.0)
min_val_2 = df['Nota'].max()
max_val_2 = df.Nota.min()
print("""
    Min_val:     %d
    Q1:          %d
    Mediana(Q2): %d
    Q3:          %d
    Max_val:     %d
""" % (min_val, Q1, mediana, Q3, max_val))
print(f'{max_val_2},  {min_val_2}')
```

## **Rango Intercuartil**

```python
#Calcular el rango intercuartil
iqr = Q3 - Q1
iqr
```

## **Datos Atipicos o Outliers**

```python
#Al calcular los datos atipicos, con esto encontrariamos el rando de los datos que consideramos anomalos
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 + 1.5*iqr
print('rango para detección de outliers: {}, {}'.format(minlimit, maxlimit))
```

# **Ejemplo:**
Supongamos que estamos analizando la estabilidad del voltaje en una red eléctrica durante un día típico. Registraremos la tensión en voltios cada hora durante 24 horas y luego calcularemos la desviación estándar para evaluar la variabilidad en el suministro eléctrico, con esto se podría detener un impacto negativo en la calidad y la estabilidad de la energía eléctrica proporcionada a los usuarios.

```python
df_1=pd.DataFrame({'Vol/hora':[120, 121, 119, 122, 120, 123, 121, 124, 122, 120, 123, 125,
                119, 121, 120, 122, 124, 121, 120, 123, 122, 120, 121, 126]})
df_1
```

## **Medidas de tendencia Central:**

```python
media = df_1['Vol/hora'].mean()
mediana = df_1['Vol/hora'].median()
moda = df_1['Vol/hora'].mode()
print(f"Media: {media} \nMediana: {mediana} \nModa: {moda} ")
```

## **Medidas de Variabilidad:**

```python
# Varianza
var = df_1['Vol/hora'].var()

# Desviación Estándar
std = df_1['Vol/hora'].std()

print(f"Varianza: {var} \nDesviación Estándar: {std}")
```

## **Interpretación:**
La media de las lecturas de voltaje es de aproximadamente **121.625 voltios**, lo que representa el nivel central de voltaje. **La desviación estándar de aproximadamente 1.86 voltios** indica cuánto varían las lecturas de voltaje en torno a esta media. En este caso, una desviación estándar más alta podría indicar una mayor variabilidad en el suministro eléctrico durante el día, lo que podría afectar la calidad y la estabilidad de la energía eléctrica proporcionada a los usuarios.

La desviación estándar es una medida importante en el suministro eléctrico porque ayuda a evaluar la consistencia y la estabilidad de la tensión suministrada, lo que es crucial para garantizar un suministro eléctrico confiable y seguro.

Si la variabilidad en el suministro eléctrico es muy alta, esto podría tener un impacto negativo en la calidad y la estabilidad de la energía eléctrica proporcionada a los usuarios. En tales casos, es importante tomar medidas para mitigar los efectos de esta variabilidad y garantizar un suministro eléctrico confiable y seguro.

**Aquí hay algunas formas en las que estos datos pueden ser útiles para generar protecciones y mejorar la calidad del suministro eléctrico:**

* **Detección de Anomalías**: Utilizando datos históricos de voltaje, puedes identificar patrones de variabilidad inusuales o anomalías. Si se detecta un voltaje que se desvía significativamente de la norma, se pueden activar alarmas o protecciones para evitar daños en equipos sensibles.

* **Protección contra Sobretensiones y Subtensiones**: La variabilidad extrema en el voltaje puede causar sobretensiones o subtensiones perjudiciales para los dispositivos eléctricos y electrónicos. Puedes instalar dispositivos de protección, como supresores de sobretensiones o reguladores de voltaje, para mantener el voltaje dentro de rangos seguros.

* **Respuesta Rápida a Cortes de Energía**: Si la variabilidad es tan alta que puede llevar a cortes de energía frecuentes, puedes implementar sistemas de respaldo, como generadores de energía de emergencia o sistemas de baterías, para garantizar la continuidad del suministro durante interrupciones.

* **Monitoreo Continuo**: El monitoreo en tiempo real de la calidad del voltaje puede ayudarte a identificar problemas rápidamente y tomar medidas correctivas antes de que causen daños. Puedes utilizar dispositivos de monitoreo de calidad de energía eléctrica para este fin.

* **Planificación de la Infraestructura**: Si la variabilidad es crónica, puede ser necesario realizar mejoras en la infraestructura eléctrica, como la expansión de la capacidad de generación o la mejora de la distribución. Los datos de variabilidad pueden ayudar en la planificación de estas mejoras.

* **Cumplimiento de Normativas**: Existen regulaciones y estándares que establecen los límites de variabilidad del voltaje que deben cumplir los proveedores de energía eléctrica. El monitoreo y el análisis de los datos pueden ayudarte a asegurarte de que se cumple con estas regulaciones.
